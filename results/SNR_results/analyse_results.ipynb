```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
def SNR (p,n=100,q = 0.1):
    return 0.5*n*(p-q)**2/(p+q)
```

```python
def SNR_small_n (p,n=100,q = 0.1):
    return n*(p-q)**2/(p+q)
```

```python
## Results from the results directory

results_BN ={}
results_BN['0.11'] = 50
results_BN['0.12'] = 50
results_BN['0.13'] = 50
results_BN['0.135'] = 67.4877
results_BN['0.14'] = 64.15 ## 64.15
results_BN['0.17'] = 79.8018
results_BN['0.3'] = 97.6701
results_BN['0.75'] =98.6067
```

```python
## Results from the results directory
results_LN = {}
results_LN[0.11] = 50
results_LN[0.12] = 50
results_LN[0.13] = 50
results_LN[0.135] = 64.2447
#results_LN[0.14] = 75
results_LN[0.17] = 74.7511
results_LN[0.3] = 90.6701
results_LN[0.75] =93.2598
```

```python
plt.figure(figsize = (15,15))

plt.plot(SNR(np.array(list(map(float,results_BN.keys())))), results_BN.values(),label = 'BN',marker = 'o',markersize = 15, lw = 4,mfc = 'red')
plt.plot(SNR(np.array(list(map(float,results_LN.keys())))), results_LN.values(),label = 'LN',marker = 'o',markersize = 15, lw = 4,mfc = 'red')


plt.title ('Test accuracy for different SNR',fontsize = 25)

plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20 )
plt.ylabel ('Test accuracy',fontsize = 25)

plt.xlabel('SNR',fontsize = 25)
plt.xscale('log',base=10) 
plt.legend(loc='best',fontsize = 25)
```

```python
plt.figure(figsize = (20,20))

plt.plot(SNR_small_n(np.array(list(map(float,results_BN.keys())))), results_BN.values(),label = 'BN',marker = 'o',markersize = 15, lw = 4,mfc = 'red')
plt.plot(SNR_small_n(np.array(list(map(float,results_LN.keys())))), results_LN.values(),label = 'LN',marker = 'o',markersize = 15, lw = 4,mfc = 'red')

plt.title ('Test accuracy for different SNR',fontsize = 25)

plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20 )
plt.ylabel ('Test accuracy',fontsize = 25)
plt.xlabel('SNR',fontsize = 25)
plt.xscale('log',base=10) 
plt.legend(loc='best',fontsize = 25)
```
